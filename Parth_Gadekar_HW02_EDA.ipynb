##Answer 1
###Importing all the necessary libraries and reading the csv file

```python
# ---------------------------------------------------
# Q1: Load dataset & Exploratory Data Analysis (EDA)
# ---------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('breast-cancer-wisconsin.csv')
df.head()
```

###I. Summarizing each column (e.g. min, max, mean )

```python
# ----------------------------
# I. Summarizing each column
# ----------------------------

summary = df.describe()
print(summary)
```

###II. Identifying missing values

```python
# --------------------------------
# II. Identifying Missing Values
# --------------------------------

missing_values = df.isna()
missing_counts = missing_values.sum()
print("Columns with missing values:")
print(missing_counts[missing_counts > 0])
```

###III. Replacing the missing values with the “mean” of the column

```python
# ------------------------------------
# III. Replace Missing Values with Mean
# ------------------------------------

column_means = df.mean(numeric_only=True)
df_filled = df.fillna(column_means)
```

###IV. Displaying the frequency table of “Class” vs. F6

```python
# ------------------------------------
# IV. Frequency Table of Class vs. F6
# ------------------------------------

frequency_table = pd.crosstab(index=df['Class'], columns=df['F6'])
print(frequency_table)
```

###V. Displaying the scatter plot of F1 to F6, one pair at a time

```python
# -------------------------------------------------
# V. Scatter Plot of F1 to F6 (One Pair at a Time)
# -------------------------------------------------

columns_to_plot = ['F1', 'F2', 'F3', 'F4', 'F5', 'F6']

for i in range(len(columns_to_plot)):
    for j in range(i + 1, len(columns_to_plot)):
        column_x = columns_to_plot[i]
        column_y = columns_to_plot[j]

        plt.figure(figsize=(8, 6))

        plt.scatter(df[column_x], df[column_y], alpha=0.5)
        plt.title(f'Scatter Plot of {column_x} vs {column_y}')

        plt.xlabel(column_x)
        plt.ylabel(column_y)

        plt.grid(False)
        plt.show()
```

###VI. Show histogram box plot for columns F7 to F9

```python
# ------------------------------------
# VI. Histogram & Box Plot for F7 to F9
# ------------------------------------

columns_to_plot = ['F7', 'F8', 'F9']
fig, axes = plt.subplots(nrows=2, ncols=len(columns_to_plot), figsize=(12, 6))

for i, column in enumerate(columns_to_plot):
    # Histogram
    axes[0, i].hist(df[column].dropna(), bins=20, alpha=0.7, color='blue', edgecolor='black')
    axes[0, i].set_title(f'Histogram of {column}')
    axes[0, i].set_xlabel(column)
    axes[0, i].set_ylabel('Frequency')

    # Box plot
    axes[1, i].boxplot(df[column].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='blue'))
    axes[1, i].set_title(f'Box Plot of {column}')
    axes[1, i].set_xlabel(column)

plt.tight_layout()
plt.show()
```
